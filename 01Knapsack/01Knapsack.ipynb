```python
# library package
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df.Discount = df.Discount.astype(int)
df.Discounted_Price = df.Discounted_Price.astype(int)
```

```python
print('\nHere is the list of 30 items to select with $1000 budget and max discount: \n')
print(df)
```

```python
discount = list(df['Discount'])
price = list(df['Discounted_Price'])
capital = 1000

col = capital + 1
row = len(discount)
dp = [[0 for _ in range(col)] for _ in range(row)]
```

```python
for item in range(row):
    for budget in range(col):
        if price[item] > budget:
            dp[item][budget] = dp[item-1][budget]
            continue
        previous_best_discount = dp[item-1][budget]
        new_best_discount = discount[item] + dp[item-1][budget - price[item]]
        dp[item][budget] = max(previous_best_discount, new_best_discount)
solutions = []
for row in dp:
    for col in row:
        solutions.append(col)
max_discount = max(solutions)
knapsack_table = pd.DataFrame(dp)     

choosen_items = []
j = capital
for i in range (len(discount)-1,0,-1):
    if dp[i][j] != dp[i-1][j]:
        choosen_items.insert(0,i)
        j -= price[i]
total_items = len(choosen_items)

df_final = df.iloc[choosen_items]

pricelist = list (df_final['Discounted_Price'])
total_price = 0
for object in pricelist:
    total_price += object
```

```python
print ('\nFrom knapsack-DP algorithm, maximum discount is: $',max_discount)
print ('\nFollowing is the Knapsack table formed using DP:\n')
print (knapsack_table)
print ('\nTotal number of items to buy for max discount is:',total_items)
print ('\nAnd we need to buy items with following indices:',choosen_items)
print ('\nTotal money used from budget of $1000 in purchase is: $',total_price)

print('\n')
print('Following is the table of purchase decision:\n')
print(df_final)
```
